Q) What is decision tree?

A decision tree is an algorithm of supervised learning which can be used for both regression and classification tasks. It has a tree like structure where the main root node branches out into decision nodes which are followed by the leaf nodes. Therefore a decision tree is a complex if else condition which helps in classifying the dataset based on its characteristics.

Decision tree can be subclassified as: classification tree and regression tree

1.   Classification tree: We use classification tree when we can classify the output into a yes or no condition. eg: on a weekend you can either decide to go out with your friends or you can stay home and relax with your family.
2.   Regression tree: We use Regression tree when we need to predict value based on the previous outcomes of the dataset. eg: we can predict the petrol prices in the upcoming years.

Q) Why do we use decision tree?

Decision trees in machine learning provide an effective method for making decisions because they lay out the problem and all the possible outcomes. Thus it enables us to analyze the possible consequences of a decision, and as an algorithm accesses more data, it can predict outcomes for future data. In machine learning, decision trees offer simplicity and a visual representation of the possibilities when formulating outcomes.


Q) Where can it be used?

Decision trees are commonly used in business for analyzing customer data and making marketing decisions, but they can also be used in fields such as medicine, finance, and machine learning.

A decision tree can be used in many conditions to select a product based on all your needs, or to train your models on how to classify different objects based on their colour, shape, size, etc. We could train it to differentiate between frutis, making a decision on purchase of a vehicle and many more...

#ENTROPY
---

Entropy is the measure of randomness. It is the amount of information needed to accurately describe the sample. Its value is high if the elements are dissimilar or random and its value is low if the elements are similar or have the same properties. Entropy’s value always lies between 0 and 1 both included.

In simpler words entropy can be defined as the amount of impurities present in the information. For example, the first set has 10 red apples the second set has 8 red and 2 green apples and the third set has 5 red apples and 5 rotten apples. In this case, the first set will have no entropy hence it is pure, the second set will have little entropy as the impurity is less whereas the third set has high impurity and the entropy will be higher at 0.5

#INFORMATION GAIN


---

We can define information gain as a measure of how much information a feature provides about a class. Information gain helps to determine the order of attributes in the nodes of a decision tree.

The main node is referred to as the parent node, whereas sub-nodes are known as child nodes. We can use information gain to determine how good the splitting of nodes in a decision tree.

As we go from the parent node to the sub node the entropy reduces from 1 towards 0. The more the entropy removed, the greater the information gain. The higher the information gain, the better the split.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Load the Wine dataset
wine = load_wine()
X = wine.data[:, :10]
Y = wine.target

# Create a Decision Tree classifier
clf = DecisionTreeClassifier()
clf.fit(X, Y)

# Plot the Decision Tree structure
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=wine.feature_names[:10], class_names=wine.target_names)
plt.title("Decision Tree Classifier")
plt.show()

```
